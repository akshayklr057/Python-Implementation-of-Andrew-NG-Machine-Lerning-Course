```python
def displayData(X):
    m, n = X.shape
    example_width = round(math.sqrt(n))   #width of each training example(digit) or image to be plotted
    example_height = round(n/example_width) #height of each training example(digit) or image to be plotted
    pad = 1  #giving padding of 1 pixel between all the images
    display_rows = math.floor(math.sqrt(m))  #number of rows to be displayed =10
    display_cols = math.ceil(m/display_rows) #number of columns to be displayed =10
    display_array = np.ones((pad + display_rows*(example_height + pad),pad + display_cols*(example_width + pad)))   # the array to be plotted, initialising it to 1 
    #display array is 211*211 matrix (200 for digits(20*10 for horizontal and vertical axis) and 11 pixels for padding between digits and boundaries of image)
    curr_ex = 0 #current example to work on
    for i in range(1, display_rows+1):
        for j in range(1, display_cols+1):
            if (curr_ex>=m):
                break
            else:
                max_val = max(abs(X[curr_ex]))  #for scaling
                display_array[pad + (i-1)*(example_height + pad): i*(example_height+pad), pad + (j-1)*(example_width + pad) : j*(example_width+pad)] = X[curr_ex].reshape((example_height,example_width))/max_val
                #display_array[1:20,1:20] = X[0].reshape((20,20))/max_val
                curr_ex = curr_ex + 1
        if (curr_ex>=m):
            break
    plt.imshow(display_array)
    plt.axis('off')
    plt.show()
```

```python
def initialiseTheta(shape):
    theta = np.zeros(shape)
    return theta
```

```python
def sigmoid(Z):
    sig = 1/(1+ np.exp(-Z))
    return sig
```

```python
def lrCostFunction(theta, X, y, reg_parameter):
    m = y.shape[0]  #number of training examples
    J =0
    grad = np.zeros(theta.shape)
    Z = np.dot(X, theta)
    print("Z shape", Z.shape)
    hx = sigmoid(Z)
    print("hx - y shape", (hx-y).shape)
    print("X shape",(X[:,0].T).shape )
#     print(hx)
    print("dot product shape", np.dot(X[:,0].T, (hx -y)))
    J = -(1/m)*(np.sum(np.multiply(y, np.log(hx)) + np.multiply((1-y), np.log(1-hx)))) + (reg_parameter/(2*m))*(np.sum(np.power(theta[1:], 2)))
    grad_unregularized = (1/m)*(np.dot(X[:,1:].T, (hx -y)))
    grad[0] = (1/m)*(np.dot(X[:,0].T, (hx -y)))
    grad[1:] = grad_unregularized + (reg_parameter/m)*(theta[1:])
    
    return (J, grad)
            
```

In this we are going to use <a href= 'https://docs.scipy.org/doc/scipy-0.13.0/reference/generated/scipy.optimize.fmin_cg.html'>fmin_cg</a> function of scipy.optimize. It works similar to <a href ='https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.optimize.fmin_tnc.html'>fmin_tnc</a>
but it computationally efficient for large number of parameters.

The theta value will be a matrix, where each row represents value for theta parameters for each model (10 different models for 10 classes).<br>
We will use for loop till K, and get each class as digit_class:-<br>
y==digit_class will check if the value of y = that particular class of for loop.<br>
**If equal then it will 1(True) to only that class and zero(False) to rest of the classes**

```python
def gradRegularization(theta, X, y, lmbda):
    m = len(y)
    temp = sigmoid(np.dot(X, theta)) - y
    temp = np.dot(temp.T, X).T / m + theta * lmbda / m
    temp[0] = temp[0] - theta[0] * lmbda / m
    return temp
```

```python
def oneVsAll(X, y, K, reg_parameter):
    X = np.hstack((np.ones((X.shape[0],1)), X))
    theta = initialiseTheta((K,X.shape[1]))
    print(theta[0].shape)
    for i in range(K):
        print("In for loop")
        digit_class = i if i else 10
        temp = opt.fmin_tnc(func = lrCostFunction, fprime=None, x0=theta[i], args = (X, (y==digit_class).flatten() , reg_parameter))
        theta[i] = temp[0]
        
        #Not working
#         theta[i]= opt.fmin_cg(func = lrCostFunction, fprime=None, x0=theta[i], args = (X, (y==digit_class).flatten() , reg_parameter))
    return theta
    
    
    
```

```python
K = len(np.unique(y))  # number of labels
```

```python
reg_parameter = 0.1

all_theta = oneVsAll(X, y, K, reg_parameter)
```

<a href= 'https://docs.scipy.org/doc/numpy/reference/generated/numpy.argmax.html'>np.argmax()</a> is used to find out the index of maximum element along the row
<br>
Each row in np.dot(X, theta.T) contains 10 entries for 10 different models, np.argmax() returns the index of corresponding model with maximum value of all 10

```python
def predictOneVsAll(all_theta, X):
    m = X.shape[0]
    num_labels = all_theta.shape[0]
    pred = np.zeros((X.shape[0],1))
    X = np.hstack((np.ones((X.shape[0],1)), X))
    
    pred = np.argmax(sigmoid(np.dot(X, all_theta.T)), axis =1) #axis = 1 row wise maximum
    pred = [e if e else 10 for e in pred]  #if index is 1 then 10 else 1-9
    
    return pred
```
